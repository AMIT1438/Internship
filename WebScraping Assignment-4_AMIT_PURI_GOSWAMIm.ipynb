## Questiong 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
#importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#Connect to the driver
driver = webdriver.Chrome(r"chromedriver.exe")
```

```python
#opening https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos page on authomated chrome browser
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
#Entering designation and location as required in the question
scr = driver.find_elements(By.XPATH,'//tr/td')

#selecting the elements from 3 to 183 which need to be scrapped in the table 
counter=1
values=[]
for i in scr:
    if(counter<3):
        counter+=1
        continue
    elif(counter>182):
        break
    else:
        values.append(i.text)
        counter+=1
    

#declaring 5 lists for storing the data to be scrapped
rank=[]

name=[]

artist=[]

upload_date=[]

views=[]
```

```python
#storing all specific elements in five lists

for i in range(0,175,6):
    if(values[i]!=' '): 
        rank.append(values[i]) 
    else: 
        rank.append('_')
        
for i in range(1,176,6):
    if(values[i]!=' '): 
        name.append(values[i]) 
    else: 
        name.append('_')   

for i in range(2,177,6):
    if(values[i]!=' '): 
        artist.append(values[i]) 
    else: 
        artist.append('_')   

for i in range(3,178,6):
    if(values[i]!=' '): 
        views.append(values[i]) 
    else: 
        views.append('_')   

for i in range(4,179,6):
    if(values[i]!=' '): 
        upload_date.append(values[i]) 
    else: 
        upload_date.append('_')

#removing element from " "
import re
count=0
for i in name:
    search=re.findall(r'"(.+?)"',i)
    name[count]=search
    count+=1

# creating data frame and displaying it 

df=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Views':views,'Upload_date':upload_date})
df
```

## 2. Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/.
    
You need to find following details:
    
A) Match title (I.e. 1st ODI)

B) Series

C) Place

D) Date

E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#Connect to the driver
driver = webdriver.Chrome(r"chromedriver.exe")

#opening https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos page on authomated chrome browser
driver.get("https://www.bcci.tv/")
```

```python
scr=driver.find_element(By.XPATH,'//a[@class="nav-link "]')
scr.click()
```

```python
#Declaring 6 empty lists to store values asked in the question
match_title=[] 
series=[]
place=[]
date_i=[]
time_i=[]
#declaring list for storing title and series combine data
values=[]
#declaring a count variable
count=0

src=driver.find_elements(By.XPATH,'//span[@class="ng-binding"]')
for i in src:
    series.append(i.text)
       
try:
    src=driver.find_elements(By.XPATH,'//div[@class="fix-place ng-binding ng-scope"]')
    for i in src:
        values.append(i.text)
        #Searching the "-"
        c=values[count].find("-")
              
        match_title.append(values[count][0:c])
        place.append(values[count][c+1:])
        count+=1
        
except NoSuchElementException:
    place.append("_")
    match_title.append("_")

src=driver.find_elements(By.XPATH,'//h5[@class="ng-binding"]')
for i in src:
    date_i.append(i.text)

src=driver.find_elements(By.XPATH,'//h5[@class="text-right ng-binding"]')
for i in src:
    time_i.append(i.text)


```

```python
# creating data frame and displaying it 

df=pd.DataFrame({'Match Title':match_title,'Series':series,'Place':place,'Date of Match':date_i,'Time of Match':time_i})
df
```
